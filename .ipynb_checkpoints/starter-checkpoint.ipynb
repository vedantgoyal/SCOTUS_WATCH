```python
from urllib.request import urlopen
import re
from collections import defaultdict
import numpy as np
```

```python
#contains all variables corresponding to a string
var_dict = defaultdict(str)

#contains all values possible for a var
var_value_dict = defaultdict(lambda: defaultdict(str))

# Grab the page from the web
f = urlopen("http://scdb.wustl.edu/documentation.php?var=chief")
html = str(f.read())


reviews = html.split('<li class="toc">')[1:]
for variable in reviews:
    str_1 = variable.split("</a>")[0].split('class="toc')
    if (re.split(".var=",str_1[0])[1].replace('" ','') == 'intro'):
        continue
    if (re.split(".var=",str_1[0])[1].replace('" ','') == 'cite'):
        continue
    var_dict[str_1[1].replace('">','')] = re.split(".var=",str_1[0])[1].replace('" ','')
```

```python
for var in var_dict.keys():
    print(str(var)+":"+str(var_dict[var]))
```

```python
# Grab the page from the web
for var in var_dict.keys():
    #print("---------------------------")
    #print(str(var)+":"+str(var_dict[var]))
    f = urlopen("http://scdb.wustl.edu/documentation.php?var="+str(var_dict[var]))
#     f = urlopen("http://scdb.wustl.edu/documentation.php?var=vote")
#     var = 'vote'
    html = str(f.read())

    if ('<strong>Values:</strong><br />' not in html):
        continue

    #print(html)
    # Extract the (30) review elements
    reviews = html.split('<strong>Values:</strong><br />')[1]
    reviews = reviews.split('<div class="toc" onClick="toggleBlock')[0]
    #print(reviews)


    from bs4 import BeautifulSoup
    soup = BeautifulSoup(reviews)
    #print(soup.text)

    final_str = soup.text.replace("\\t","#").replace("\\n","#")
    #print(final_str)
    list_val = re.split("#+",final_str)
    list_val = [x for x in list_val if x not in [""," "]]
    #print(len(list_val))
#     if var_dict[var] == 'justice':
#         print(list_val[0])
#         print(list_val[1])
#         print(list_val[2])
#         print(list_val[3])
#         print(list_val[4])
#         print(list_val[5])
#         print(list_val)
    
    assert(len(list_val)%2 == 0 or len(list_val)%3 == 0)
    
    #choose which format it is
    pattern = re.compile("\d+")
    
    if (len(list_val)>4):
        if (pattern.match(list_val[0]) and pattern.match(list_val[2]) and pattern.match(list_val[4])):
            div = 2
        elif (pattern.match(list_val[0]) and pattern.match(list_val[3]) and pattern.match(list_val[6])):
            div = 3
        else:
            print("diff pattern found")
    elif (len(list_val)<4):
        if (len(list_val)%2==0):
            div = 2
        elif (len(list_val)%2==0):
            div = 3
    #print(div)
    
    if (div == 2):
        for x in np.arange(0,len(list_val)//2):
            #print("var_value_dict["+str(var_dict[var])+"]["+str(list_val[x*2])+"] = "+str(list_val[(x*2)+1]))
            var_value_dict[str(var_dict[var])][list_val[x*2]] = str(list_val[(x*2)+1])
    elif (div == 3):
        for x in np.arange(0,len(list_val)//3):
            #print("var_value_dict["+str(var_dict[var])+"]["+str(list_val[x*3])+"] = "+str(list_val[(x*3)+1])+ str(list_val[(x*3)+2]))
            var_value_dict[str(var_dict[var])][list_val[x*3]] = str(list_val[(x*3)+1])#+";"+str(list_val[(x*3)+2])
```
